```python
#extraer datos web
import urllib
import urllib.request
from bs4 import BeautifulSoup
```

```python
def make_soup(url):
    thepage = urllib.request.urlopen(url)
    soupdata = BeautifulSoup(thepage, "html.parser")
    return soupdata
```

```python
url = "https://www.camara.cl/diputados/diputados.aspx#mostrarDiputados"
```

```python
soup = make_soup(url)
```

```python
print(soup)
```

```python
tabla=soup.find_all('div')
```

```python
print(tabla)
```

```python
tabla_diputados = soup.find("div", {"id": "ContentPlaceHolder1_ContentPlaceHolder1_pnlDiputadosLista"})
```

```python
print(tabla_diputados)
```

```python
lista = tabla_diputados.find_all('article')
```

```python
lista[10]
```

```python
lista[10].find('img')
```

```python
lista[10].find('img').get('title')
```

```python
lista[0].find_all('p')[1].text
```

```python
diputados = []
partidos = []

for persona in lista:
    diputados.append(persona.find('img').get('title'))
    partidos.append(persona.find_all('p')[1].text)
    
```

```python
diputados
```

```python
partidos
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(diputados, partidos)), 
               columns =['senador', 'partido']) 
df.head() 
```

```python
df.to_csv('diputados.csv')
```
